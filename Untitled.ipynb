# Первое задание

```python
import pandas as pd 
import numpy as np 
import sklearn.datasets as dt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn import preprocessing
from prettytable import PrettyTable


count = 150
```

```python

X, y = dt.make_classification(
    n_samples=count, 
    n_features=2, 
    n_redundant=0,
    n_informative=2, 
    class_sep=0.1,
    n_classes=2, 
    random_state=1 
)
dataset1 = pd.DataFrame(X)
dataset1.columns = ['x', 'y']
dataset1['target'] = y

data1 = []
target1 = []


for i in range(count):
    data1.append([dataset1['x'][i], dataset1['y'][i]])
    target1.append(dataset1['target'][i])
    
data1_train, data1_test, target1_train, target1_test = train_test_split(data1, target1, test_size=0.2, random_state=0)

k1 = []
accuracy1 = []
for i in range(1, 9):
    knn = KNeighborsClassifier(n_neighbors = i)# Fit the classifier to the data
    knn.fit(data1_train,target1_train)
    plt.xlabel('k')
    plt.ylabel('accuracy')
    k1.append(i)
    accuracy1.append(knn.score(data1_test, target1_test))
plt.plot(k1,accuracy1)
plt.show()

scaler = preprocessing.StandardScaler().fit(data1_train)
data1_train = scaler.transform(data1_train)
data1_test = scaler.transform(data1_test)
target1_pred = knn.predict(data1_test)

print("Точность: ", accuracy_score(target1_test, target1_pred))
print("F1-мера",f1_score(target1_test, target1_pred))
print("Полнота",recall_score(target1_test, target1_pred))

```

# Второе задание

```python
C="Cat"
F="Fish"
H="Hen"
y_true = [C,C,C,C,C,C, F,F,F,F,F,F,F,F,F,F, H,H,H,H,H,H,H,H,H]
y_pred = [C,C,C,C,H,F, C,C,C,C,C,C,H,H,F,F, C,C,C,H,H,H,H,H,H]
p, r, f1 = [], [], []
ip= lp= io= lo= s_ip= s_lp= s_io= s_lo = 0
s = [y_true.count(C), y_true.count(F), y_true.count(H)]
j = 0
for var in [C, F, H]:
    for i in range(len(y_pred)):
        if y_true[i]==var and y_pred[i] == var:
            ip+=1
            s_ip+=1
        elif y_true[i]==var and y_pred[i] != var:
            lo+=1
            s_lo+=1
        elif y_true[i]!=var and y_pred[i] != var:
            io+=1
            s_io+=1
        else:
            lp+=1
            s_lp+=1
    p.append(round(ip/(ip+lp), 3))
    r.append(round(ip/(ip+lo), 3))
    f1.append(round(1/(1/p[j]+ 1/r[j]), 3))
    j+=1
    ip= lp= io= lo = 0
accuracy, macro_avg, weighted_avg = [], [], []
sm1= sm2= sm3 = 0 
for i in range(3):
    sm1+=p[i]*s[i]
    sm2+=r[i]*s[i]
    sm3+=f1[i]*s[i]

accuracy.append(round((sum(p)/3), 3))
accuracy.append(round((sum(r)/3), 3))
accuracy.append(round((sum(f1)/3), 3))

macro_avg.append(round(sm1/sum(s), 3))
macro_avg.append(round(sm2/sum(s), 3))
macro_avg.append(round(sm3/sum(s), 3))

weighted_avg.append(round(s_ip/(s_ip+s_lp), 3))
weighted_avg.append(round(s_ip/(s_ip+s_lo), 3))
weighted_avg.append(round(1/(1/weighted_avg[0] + 1/weighted_avg[1]), 3))

x = PrettyTable()

x.field_names = [" ", "Precision", "Recall", "F1-score", "Support"]

x.add_row(["Cat", p[0], r[0], f1[0], s[0]])
x.add_row(["Fish", p[1], r[1], f1[1], s[1]])
x.add_row(["Hen", p[2], r[2], f1[2], s[2]])
x.add_row(["Accuracy", accuracy[0], accuracy[1],accuracy[2], sum(s)])
x.add_row(["Macro_avg", macro_avg[0], macro_avg[1], macro_avg[2], sum(s)])
x.add_row(["Weighted_avg", weighted_avg[0], weighted_avg[1], weighted_avg[2], sum(s)])

print(x)


              


              
              
    
```
